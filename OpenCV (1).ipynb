## Reading Images using OpenCV

Function Used - cv2.imread('image_name',flag) 

## Flags:
1. cv2.IMREAD_GRAYSCALE or 0 : Loads Grayscale mode
2. cv2.IMREAD_COLOR or 1 : Loads Color image. Default Mode
3. cv2.IMREAD_UNCHANGED or -1: Loads image as such including Alpha    Channel

```python
# Reading image as Gray Scale
gs_image=cv2.imread('checkerboard_18x18.png',0)

#Print image data- the Pixel Values, elements of a 2D Numpy array
#Each Pixel value is 8-bits [0,255]
print(gs_image)
```

```python
# Image Attributes
print("Image size is --> ",gs_image.shape)
print("Data Type of Image --> ",gs_image.dtype)
```

## Displaying Images using Matplotlib

```python
plt.imshow(gs_image)
```

Since Grayscale colormap is not set, matplotlib by default displays images in color.

```python
plt.imshow(gs_image,cmap='gray')
```

# Splitting and Merging Color channels

### cv2.split() -  Divides a Multi channel array into several single chanel arrays.

### cv2.merge() - Vice-versa. All the input channels must have same size

```python
# Split image into B,G and R Components
image_bgr=cv2.imread("RGB.png",1)
plt.imshow(image_bgr)
b,g,r=cv2.split(image_bgr)
```

```python
# Show the Channels
plt.figure(figsize=[30,10])
plt.subplot(151); plt.imshow(r,cmap='gray');plt.title("Red Channel");
plt.subplot(152); plt.imshow(g,cmap='gray');plt.title("Green Channel");
plt.subplot(153); plt.imshow(b,cmap='gray');plt.title("Blue Channel");

# Merge the channels into R,G,B
merged_image=cv2.merge((r,g,b))
plt.subplot(154);plt.imshow(merged_image);plt.title('Merged Output')
```
